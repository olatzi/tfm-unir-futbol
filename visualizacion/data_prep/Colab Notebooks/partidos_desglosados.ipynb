Lee el archivo CSV y crea un DataFrame

```python
temporadas = [
    '11-12',
    '12-13',
    '13-14',
    '14-15',
    '15-16',
    '16-17',
    '17-18',
    '18-19',
    '19-20',
    '20-21',
    '21-22'
]
```

```python
numeroTemporada = temporadas[1]
```

```python
ruta = '/content/datos_' + numeroTemporada + '.csv'
```

```python
df = pd.read_csv(ruta)
```

```python
#df = df.assign(Season=numeroTemporada)
df.insert(0, 'Season', numeroTemporada)
```

```python
# Crear la nueva columna "MatchID" en el DataFrame "df"
#df['MatchID'] = df.index + 1
df.insert(0, 'MatchID', df.index + 1)
# Imprimir el DataFrame actualizado
print(df)
```

```python
df_transformed = pd.DataFrame(columns=['MatchID', 'Date', 'Team', 'Goals', 'TeamLocation', 'OpposingTeam', 'FullTimeResult', 'HalfTimeResult',
                                     'HalfTimeGoals', 'ShotsOnTarget', 'Corners', 'FoulsCommitted', 'YellowCards', 'RedCards', 'Season'
                                     ])
```

```python
# Recorre las filas del DataFrame
for index, row in df.iterrows():
    # Accede a los valores de cada columna en la fila actual

    match_id = index+1
    date = row['Date'] # Match Date (dd/mm/yy)
    #time = row['Time']

    home_team = row['HomeTeam'] 
    away_team = row['AwayTeam']

    fullTimeResult = row['FTR'] # Full Time Result (H=Home Win, D=Draw, A=Away Win)
    halfTimeResult = row['HTR'] # Half Time Result (H=Home Win, D=Draw, A=Away Win)

    ### (H=Home Win, D=Draw, A=Away Win) ###

    if fullTimeResult == 'H':
      fullTimeResultHome = "Victoria"
      fullTimeResultAway = "Derrota"
    elif fullTimeResult == 'D':
      fullTimeResultHome = "Empate"
      fullTimeResultAway = "Empate"
    elif fullTimeResult == 'A':
      fullTimeResultHome = "Derrota"
      fullTimeResultAway = "Victoria"
    else:
      fullTimeResultHome = "Resultado desconocido"
      fullTimeResultAway = "Resultado desconocido"

    if halfTimeResult == 'H':
      halfTimeResultHome = "Victoria"
      halfTimeResultAway = "Derrota"
    elif halfTimeResult == 'D':
      halfTimeResultHome = "Empate"
      halfTimeResultAway = "Empate"
    elif halfTimeResult == 'A':
      halfTimeResultHome = "Derrota"
      halfTimeResultAway = "Victoria"
    else:
      halfTimeResultHome = "Resultado desconocido"
      halfTimeResultAway = "Resultado desconocido"


    home_goals = row['FTHG'] # Full Time Home Team Goals
    away_goals = row['FTAG'] # Full Time Away Team Goals

    hthg = row['HTHG'] # Half Time Home Team Goals -  HalfTimeGoals
    hs = row['HS'] # Home Team Shots - Shots
    hst = row['HST'] # Home Team Shots on Target -    ShotsOnTarget
    #hhw = row['HHW'] # Home Team Hit Woodwork (El equipo local golpea la madera)
    hc = row['HC'] # Home Team Corners -              Corners
    hf = row['HF'] # Home Team Fouls Committed -      FoulsCommitted
    #hfkc = row['HFKC'] # Home Team Free Kicks Conceded-------------------------
    #ho = row['HO'] # Home Team Offsides----------------------------------------
    hy = row['HY'] # Home Team Yellow Cards -         YellowCards
    hr = row['HR'] # Home Team Red Cards -            RedCards
    #hbp = row['HBP'] # Home Team Bookings Points (10 = yellow, 25 = red)-------
    
    ###########  AWAY  #########################################################
    htag = row['HTAG'] # Half Time Away Team Goals - HalfTimeGoals
    as_ = row['AS'] # Away Team Shots - Shots
    ast = row['AST'] # Away Team Shots on Target - ShotsOnTarget
    #ahw = row['AHW'] # Away Team Hit Woodwork----------------------------------
    ac = row['AC'] # Away Team Corners - Corners
    af = row['AF'] # Away Team Fouls Committed - FoulsCommitted
    #afkc = row['AFKC'] # Away Team Free Kicks Conceded-------------------------
    #ao = row['AO'] # Away Team Offsides ---------------------------------------
    ay = row['AY'] # Away Team Yellow Cards -         YellowCards
    ar = row['AR'] # Away Team Red Cards -            RedCards
    #abp = row['ABP'] # Away Team Bookings Points (10 = yellow, 25 = red)-------


    # Crear la primera fila para el equipo local
    row_local = pd.DataFrame({
          'MatchID': [match_id], 'Date': [date], 'Team': [home_team], 'Goals': [home_goals], 'TeamLocation': "Local", 'OpposingTeam': [away_team], 'FullTimeResult': [fullTimeResultHome], 'HalfTimeResult': [halfTimeResultHome],   
          'HalfTimeGoals': [hthg], 'ShotsOnTarget': [hst], 'Corners': [hc], 'FoulsCommitted': [hf], 'YellowCards': [hy], 'RedCards': [hr], 'Season': numeroTemporada
          })
    
    # Crear la segunda fila para el equipo visitante
    row_visitor = pd.DataFrame({
          'MatchID': [match_id], 'Date': [date], 'Team': [away_team], 'Goals': [away_goals], 'TeamLocation': "Visitante", 'OpposingTeam': [home_team], 'FullTimeResult': [fullTimeResultAway], 'HalfTimeResult': [halfTimeResultAway],   
          'HalfTimeGoals': [htag], 'ShotsOnTarget': [ast], 'Corners': [ac], 'FoulsCommitted': [af], 'YellowCards': [ay], 'RedCards': [ar], 'Season': numeroTemporada
          })
    
    # Concatenar las filas al DataFrame df_transformed
    df_transformed = pd.concat([df_transformed, row_local, row_visitor], ignore_index=True)

# Convertir las columnas 'MatchID' y 'Goals' a tipo numérico
df_transformed['MatchID'] = pd.to_numeric(df_transformed['MatchID'])
df_transformed['Goals'] = pd.to_numeric(df_transformed['Goals'])
```

```python
# Imprimir el nuevo DataFrame df_transformed
print(df_transformed)
```
