```python
import pandas as pd
```

```python
temporadas = [
    '11-12',
    '12-13',
    '13-14',
    '14-15',
    '15-16',
    '16-17',
    '17-18',
    '18-19',
    '19-20',
    '20-21',
    '21-22'
]
```

```python
dfClasificaciones11_22 = pd.DataFrame()
```

```python
for temporada in temporadas:
    ruta = '/content/clasificaciones' + temporada + '.csv'
    df = pd.read_csv(ruta)
    # Agregar el DataFrame de la temporada actual al DataFrame general
    dfClasificaciones11_22 = pd.concat([dfClasificaciones11_22, df], ignore_index=True)
```

```python
dfClasificaciones11_22
```

### Guardar CSV con todas las clasificaciones del 2011 al 2022

```python
nombre_csv = 'clasificacionesLaLiga_MERGE_2011-2022.csv'
```

```python
dfClasificaciones11_22.to_csv(nombre_csv, index=False)
```
