```python
import pandas as pd
```

```python
temporadas = [
    '11-12',
    '12-13',
    '13-14',
    '14-15',
    '15-16',
    '16-17',
    '17-18',
    '18-19',
    '19-20',
    '20-21',
    '21-22'
]
```

```python
dfPartidosDesglosados11_22 = pd.DataFrame()
```

```python
for temporada in temporadas:
    ruta = '/content/partidos_desglosados' + temporada + '.csv'
    df = pd.read_csv(ruta)
    # Agregar el DataFrame de la temporada actual al DataFrame general
    dfPartidosDesglosados11_22 = pd.concat([dfPartidosDesglosados11_22, df], ignore_index=True)
```

```python
dfPartidosDesglosados11_22
```

```python
for columna in dfPartidosDesglosados11_22.columns:
  
  print(columna)
```

### Guardar CSV con todas las clasificaciones del 2011 al 2022

```python
nombre_csv = 'ParidosDesglose_LaLiga_MERGE_2011-2022.csv'
```

```python
dfPartidosDesglosados11_22.to_csv(nombre_csv, index=False)
```
